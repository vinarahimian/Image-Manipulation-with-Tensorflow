```python
mean = 0.0
sigma = 1.0
```

```python
#The formula for Gaussian Curve (got it from wikipedia)
z = (tf.exp(tf.negative(tf.pow(x - mean, 2.0) /
                   (2.0 * tf.pow(sigma, 2.0)))) *
     (1.0 / (sigma * tf.sqrt(2.0 * 3.1415))))
```

```python
z 
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

plt.plot(x.eval(), z.eval())
```
